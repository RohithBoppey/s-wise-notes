```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

input_image = cv2.imread('Sample_3.tif')
```

```python
# Function to show the grayscaled image
def showGrayImage(image, title):
    plt.figure(figsize=(9, 9))
    plt.imshow(image, cmap=plt.cm.gray)
    plt.title(title)
    plt.show()

# Function used to show the original input
def showOriginalImage(image, title):
    # print(title)
    plt.figure(figsize=(9, 9))
    plt.imshow(image)
    plt.title(title)
    plt.show()

# This function normalizes the pixels
def normalize_pixels(pixels):
    min = pixels.min()
    max = pixels.max()
    pixels -= min
    pixels = pixels.astype(float)
    pixels *= (255/(max-min))
    return pixels.astype(np.uint8)
```

```python
# showing the original image
print('Original image is as follows: ')
showOriginalImage(input_image, 'Original Input Image')
```

```python
# using Kernel
kernel = np.array([[0, 1, 0], [1, -4, 1], [0, 1, 0]])
laplacian_filter = cv2.filter2D(input_image, -1, kernel)
laplacian_filter = normalize_pixels(laplacian_filter)
gray_image = cv2.cvtColor(laplacian_filter, cv2.COLOR_BGR2GRAY)
```

```python
# After adding Laplacian filter
showGrayImage(gray_image, 'Laplacian of the Given Input Image')
```

```python
# Normalizing the result
normalized_image = normalize_pixels(input_image+laplacian_filter)
showOriginalImage(normalized_image,
                  'Sharpening after adding Laplacian')
```

```python
# For Sobel gradient
kernel = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
sobel_gradient = cv2.filter2D(input_image, -1, kernel)
sobel_gradient = normalize_pixels(sobel_gradient)

# Sobel gradient image
showOriginalImage(sobel_gradient, "Sobel gradient of Image")
```

```python
kernel = np.ones((5, 5))/25
smooth_sobel = cv2.filter2D(sobel_gradient, -1, kernel)

showOriginalImage(
    smooth_sobel, "Sobel Image after Smoothing (5x5 filter)")
```

```python
# Masking happens
key = 1
mask_filter = smooth_sobel + key * (input_image + laplacian_filter)
mask_filter = normalize_pixels(mask_filter)

showOriginalImage(
    mask_filter, "Masked Image formed after Laplacian filter * Sobel Smooth filer")
```

```python
# Adding mask to the input image
mask_complete = input_image + mask_filter
mask_complete = normalize_pixels(mask_complete)

showOriginalImage(
    mask_complete, "Sharpened image after sum of Image and Mask filter")
```

```python
# Applying power law transformation
power_law_transformer = np.array(255*(mask_complete/255)**0.5, dtype='uint8')
power_law_transformer = normalize_pixels(power_law_transformer)
```

```python
showOriginalImage(
    power_law_transformer, "Power-law transformed image from Sharpened Image")
```
